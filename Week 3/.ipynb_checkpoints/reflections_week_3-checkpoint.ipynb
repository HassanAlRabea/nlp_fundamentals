```python
# Remove the uneeded columns & filter out alpha & stop words
df_cleaned = df[(df['is_alpha'] == True) & (df['stop_word'] == False)]
df_cleaned = df_cleaned.drop(columns=['filename', 'content', 'is_alpha', 'stop_word'], axis=1)
# also lowercase words
df_cleaned["word"] = df_cleaned["word"].apply(lambda x: x.lower())
df_cleaned['wordid'] = df_cleaned.reset_index().index
df_cleaned
```

```python
# Analyzing with the NRC Lexicon using NRCLex
from nrclex import NRCLex

df_cleaned["NRC"] = df_cleaned["word"].apply(lambda x: NRCLex(x).affect_frequencies)
df_cleaned
```

```python
# Splitting the dict to columns
df_cleaned_nrc = pd.json_normalize(df_cleaned.NRC)
df_cleaned_nrc['wordid'] = df_cleaned_nrc.reset_index().index
df_cleaned_nrc
```

```python

df_cleaned = df_cleaned.set_index('wordid').join(df_cleaned_nrc.set_index('wordid'))
df_cleaned
```

```python
df_cleaned['is_nrc']= df_cleaned.iloc[:, -12:-1].sum(axis=1)
df_cleaned[df_cleaned['is_nrc'] > 0]
```

```python
#Now with afinn
from afinn import Afinn
#instantiate afinn
afn = Afinn()

df_cleaned["AFINN"] = df_cleaned["word"].apply(lambda x: afn.score(x))
df_cleaned["afinn_sentiment"] = df_cleaned["AFINN"].apply(lambda x: 'positive' if x > 0 else 'negative' if x < 0 else 'neutral')
df_cleaned
```

```python
#Now with Bing

with open('negative-words.txt') as f:
    negative_words = [line.rstrip('\n') for line in f]
    
with open('positive-words.txt') as f:
    positive_words = [line.rstrip('\n') for line in f]

df_cleaned["BING"] = df["word"].apply(lambda x: 'positive' if x in positive_words else 'negative' if x in negative_words else 'neutral')
df_cleaned
```

```python
# Results
print(df_cleaned[df_cleaned["is_nrc"] > 0].count())
print(df_cleaned[df_cleaned["AFINN"] != 0].count())  # Assume neutral means it doesnt exist
print(df_cleaned[df_cleaned["BING"] != 'neutral'].count())  # Assume neutral means it doesnt exist
```
