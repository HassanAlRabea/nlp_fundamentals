```python
# Importing the BBC News dataset
import numpy as np
import pandas as pd

df = pd.read_csv('bbc-news-data.csv', sep='\t')

# The df is split into four columns, where each row is an article. 
df.head()
```

```python
df.loc[0, "content"]
```

```python
# Creating a copy of the original DF, where we want the rows to be split by sentence instead of article
df_sen = df.copy()
df_sen
```

```python
# Creating the regEX to split the text by sentences
import re

rule_sen = r"(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?|!)\s{1,}"
df_sen["sentence"] = df_sen["content"].apply(lambda x: re.split(rule_sen, x))
```

### Breaking the logic down for sentence extracation:
- Made up of 3 capturing groups

- (?<!\w\.\w.) negative lookbehind where it tries to match for (?<![A-Z][a-z]\.), which is any lower case or uppercase letters followed by a period, and where the preceding characters is not  any word charchter followed by a period \w\.\w. (i.e. MATCH)

- (?<![A-Z][a-z]\.) negative lookbehind where it tried to match for !, ? or . that is not preceeded by any lowercase or uppercase letter following with dot character. (Mr. MATCH) 

- (?<=\.|\?|!)\s positive lookbehind where it tries to match for \s (any whitespace charachter) preceeded by a period, a question mark or an exlamation mark (how we actually end sentences)

- To sum up, weâre searching for whitespace characters by checking its preceding characters. If it passes from our three validations, we match it. So, we can split the text to sentences from this.

```python
#Exploding the DF by the new sentence column instead of by text
df_sen = df_sen.explode("sentence", ignore_index=True)
df_sen.rename(columns={"Unnamed: 0": "sentence id"}, inplace=True)
df_sen.index.name = "sentence id"
df_sen
```

```python
#Exploding the DF by the new paragraph column instead of by text
df_par = df_par.explode("paragraph", ignore_index=True)
df_par.rename(columns={"Unnamed: 0": "paragraph id"}, inplace=True)
df_par.index.name = "paragraph id"
df_par
```

```python
## We should split the texts by words first. 
## Turning the pandas content column into a string

raw_text = pd.Series(df['content'], dtype="string", index=None)
raw_text[0]
```

```python
# Copying dataset to split by names
df_name = df.copy()

#THe rule for paragraphs is to simply split the text whenevr there is 2 or more whitespaces 
rule_name = r"\b[A-Z][A-Za-z]+\b"
df_name["names"] = df_name["content"].apply(lambda x: re.findall(rule_name, x))
raw_names = pd.Series(df_name["names"], dtype="string", index=None)
raw_names
```

```python
# After extracting the list of names, we process it further and count the number of names and frequerncy

joined_list =' '.join(raw_names)
# creates tokens, creates lower class, removes numbers & stopwords
name_tokens = word_tokenize(joined_list)
name_tokens = [re.sub(r'\'', '', t) for t in name_tokens if re.match(r"\'\b[A-Za-z]+\b", t)]
name_tokens = [t.lower() for t in name_tokens]
# # Added the additional stop words to the list
name_tokens =[t for t in name_tokens if t not in all_stop_words]
name_tokens = [t for t in name_tokens if t.isalpha()]
# #Counts the words
counted = Counter(name_tokens)
word_freq_name = pd.DataFrame(counted.items(),columns=['word','frequency']).sort_values(by='frequency',ascending=False)
word_freq_name
```

```python

def word_frequency(sentence):
    # joins all the sentenses
    sentence ="".join(sentence)
    # creates tokens, creates lower class, removes numbers & stopwords
    new_tokens = word_tokenize(sentence)
    new_tokens = [t.lower() for t in new_tokens]
    # Added the additional stop words to the list
    new_tokens =[t for t in new_tokens if t not in all_stop_words]
    new_tokens = [t for t in new_tokens if t.isalpha()]
    #Counts the words
    counted = Counter(new_tokens)
    word_freq = pd.DataFrame(counted.items(),columns=['word','frequency']).sort_values(by='frequency',ascending=False)
    return word_freq


def word_frequency_with_stop(sentence):
    # joins all the sentenses
    sentence =" ".join(sentence)
    # creates tokens, creates lower class, removes numbers & stopwords
    new_tokens = word_tokenize(sentence)
    new_tokens = [t.lower() for t in new_tokens]
    new_tokens = [t for t in new_tokens if t.isalpha()]
    #Counts the words
    counted = Counter(new_tokens)
    word_freq = pd.DataFrame(counted.items(),columns=['word','frequency']).sort_values(by='frequency',ascending=False)
    return word_freq
```

```python
# Finding word frequency on all 
word_freq = word_frequency(raw_text)
word_freq_w_stop = word_frequency_with_stop(raw_text)
```

```python
# Finding word frequency of names
word_freq_names = word_frequency(raw_names)
word_freq_names
```
